```python
#import all the necessary libraries 
import pygame
import random
import time
```

```python
# Initialize pygame and font
pygame.init()
pygame.font.init()
```

```python
# Set up display
WIDTH, HEIGHT = 600, 400
window = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Rock Paper Scissors Game")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
BLUE = (0, 0, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)

# Load images
rock_img = pygame.image.load('rock.png')
paper_img = pygame.image.load('paper.png')
scissors_img = pygame.image.load('scissors.png')

# Resize images
rock_img = pygame.transform.scale(rock_img, (100, 100))
paper_img = pygame.transform.scale(paper_img, (100, 100))
scissors_img = pygame.transform.scale(scissors_img, (100, 100))

# Font
font = pygame.font.Font(None, 50)

# Game variables
player_choice = None
computer_choice = None
choices = ['rock', 'paper', 'scissors']
result_text = ""
confetti = []
confetti_duration = 0  # Duration for confetti

# Set up FPS control
clock = pygame.time.Clock()

# Function to draw text
def draw_text(text, font, color, surface, x, y):
    text_obj = font.render(text, True, color)
    text_rect = text_obj.get_rect(center=(x, y))
    surface.blit(text_obj, text_rect)
```

```python
# Function to check for winner
def get_winner(player, computer):
    if player == computer:
        return "Draw!"
    elif (player == 'rock' and computer == 'scissors') or \
         (player == 'paper' and computer == 'rock') or \
         (player == 'scissors' and computer == 'paper'):
        return "You Win!"
    else:
        return "You Lose!"

# Confetti creation for winning
def create_confetti():
    confetti_rect = pygame.Rect(random.randint(0, WIDTH), 0, 5, 15)
    confetti_color = random.choice([RED, GREEN, BLUE, (255, 255, 0), (255, 165, 0), (138, 43, 226)])
    confetti.append((confetti_rect, confetti_color))

# Confetti falling animation
def draw_confetti():
    for rect, color in confetti:
        pygame.draw.rect(window, color, rect)
        rect.move_ip(0, random.randint(5, 10))  # Make the confetti fall
        if rect.top > HEIGHT:
            confetti.remove((rect, color))  # Remove confetti when it falls off-screen

# Game loop
running = True
while running:
    window.fill(BLACK)  # Change background to BLACK

    # Draw game options
    window.blit(rock_img, (50, 150))
    window.blit(paper_img, (250, 150))
    window.blit(scissors_img, (450, 150))
    
    # Draw text
    draw_text("Choose: Rock, Paper, Scissors", font, WHITE, window, WIDTH//2, 50)  # Adjust text color for visibility
    
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        if event.type == pygame.MOUSEBUTTONDOWN:
            x, y = event.pos
            
            # Player selection logic
            if 50 < x < 150 and 150 < y < 250:
                player_choice = 'rock'
            elif 250 < x < 350 and 150 < y < 250:
                player_choice = 'paper'
            elif 450 < x < 550 and 150 < y < 250:
                player_choice = 'scissors'
            
            # Computer random choice
            if player_choice:
                computer_choice = random.choice(choices)
                result_text = get_winner(player_choice, computer_choice)

    # Show the result if both player and computer have made choices
    if player_choice and computer_choice:
        result_color = GREEN if result_text == "You Win!" else RED
        draw_text(f"Player: {player_choice}", font, WHITE, window, WIDTH//2, 300)
        draw_text(f"Computer: {computer_choice}", font, WHITE, window, WIDTH//2, 350)
        draw_text(result_text, font, result_color, window, WIDTH//2, 380)

        # Confetti effect when the player wins
        if result_text == "You Win!":
            confetti_duration = 300  # Set confetti duration to last for 5 seconds (300 frames)
            for _ in range(100):  # Generate a bunch of confetti
                create_confetti()
        
        # Wait for a short period and reset
        pygame.display.update()
        time.sleep(2)
        player_choice, computer_choice, result_text = None, None, ""

    # Draw confetti if it exists and confetti duration is active
    if confetti_duration > 0:
        draw_confetti()
        confetti_duration -= 1  # Decrease the duration with each frame
    else:
        confetti.clear()  # Clear confetti after duration ends

    # Update display
    pygame.display.update()
    
    # Control frame rate
    clock.tick(60)  # Ensure 60 FPS

# Quit the game
pygame.quit()
```
